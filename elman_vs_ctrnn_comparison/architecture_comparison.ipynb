# Architectural Comparison: Elman RNN vs. Continuous-Time RNN (CTRNN)

**Hypothesis:** Standard discrete-time recurrent networks (like the Elman RNN) will fail to learn high-variance integration tasks because they track instantaneous noise fluctuations. A Continuous-Time RNN (CTRNN) equipped with a biological time constant ($\tau$) acts as a low-pass filter, enabling stable accumulation of noisy evidence and successful task convergence.

To justify our architectural choice for the main Partial Information Decomposition (PID) analysis, we train both an Elman RNN and a CTRNN (hidden size $N=100$) on the Mante et al. (2013) Perceptual and Context tasks. We evaluate not only their task accuracy but also whether they possess the stable representational capacity required for information geometry analysis.

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
import copy
import os
import json

#import sys
#sys.path.append('../src/tasks') # Ensure Python can find data_loader.py
from src.tasks.data_loader import load_mante_data
from src.models.ctrnn import CTRNN
from src.models.RNN import ElmanRNN
from src.analysis.gaussian_pid import gaussian_pid_rnn_gpu
```

```python
# Configuration
BASE_PATH = "../data"
BATCH_SIZE = 1024 * 4
SUBSAMPLE_STEP = 1 # Subsample for Elman RNN (simulate tau=10ms)

print("Loading Perceptual Datasets...")
per_train_loader = load_mante_data(f'{BASE_PATH}/perceptual/train.npz', batch_size=BATCH_SIZE, shuffle=True, subsample_step=SUBSAMPLE_STEP)
per_val_loader = load_mante_data(f'{BASE_PATH}/perceptual/val.npz', batch_size=BATCH_SIZE, shuffle=False, subsample_step=SUBSAMPLE_STEP)

print("Loading Context Datasets...")
ctx_train_loader = load_mante_data(f'{BASE_PATH}/context/train.npz', batch_size=BATCH_SIZE, shuffle=True, subsample_step=SUBSAMPLE_STEP)
ctx_val_loader = load_mante_data(f'{BASE_PATH}/context/val.npz', batch_size=BATCH_SIZE, shuffle=False, subsample_step=SUBSAMPLE_STEP)

print("All datasets loaded successfully!")
```

### 1. Model Architectures
We define wrappers for both the Elman RNN and the CTRNN. 
* **Elman RNN:** Updates its hidden state entirely at every timestep $t$. Highly susceptible to the $\sigma=1.0$ stimulus noise.
* **CTRNN:** Updates its hidden state using an Euler integration step scaled by $\Delta t / \tau$ (where $\tau=100$ ms and $\Delta t=10$ ms). This biological "leak" forces the network to retain a memory of the previous state, naturally smoothing out the noise.

```python
class WrappedCTRNN(nn.Module):
    def __init__(self, input_dim, hidden_size, output_size=3):
        super().__init__()
        # Use the existing CTRNN
        self.model = CTRNN(input_size=input_dim, hidden_size=hidden_size, output_size=output_size)
        
    def forward(self, x):
        # CTRNN expects (Batch, Seq, Dim). 
        # We grab outputs and hidden_states, ignoring predictions.
        outputs, _, hidden_states = self.model(x, return_dynamics=True)
        return outputs, hidden_states

class WrappedElman(nn.Module):
    def __init__(self, input_dim, hidden_size, output_size=3):
        super().__init__()
        # Use ElmanRNN, override to 80 units
        self.model = ElmanRNN(dim=input_dim, hidden_dim=hidden_size)
        # Hardcoded output=input_dim. We overwrite it to 3 for our task classes.
        self.model.readout = nn.Linear(hidden_size, output_size) 
        
    def forward(self, x):
        # x is (Batch, Seq, Dim). 
        # ElmanRNN forward() assumes batch=1 and seq_first.
        # We bypass the forward() and just call the internal layers directly to support batching.
        x_seq_first = x.transpose(0, 1) # Convert to (Seq, Batch, Dim)
        
        h_seq, _ = self.model.rnn(x_seq_first) 
        outputs = self.model.readout(h_seq)
        
        # Transpose back to (Batch, Seq, Features) for the training loop
        return outputs.transpose(0, 1), h_seq.transpose(0, 1)
```

### 2. Loss Function
Because the network only needs to output a valid choice during the final decision period (ignoring the fixation period), we use a masked Cross-Entropy loss. This prevents the optimizer from artificially penalizing the network while it is still accumulating evidence.

```python
def masked_cross_entropy(outputs, targets):
    """Computes loss only on the decision timesteps (where target != 0)"""
    mask = targets != 0
    if mask.sum() == 0:
        return torch.tensor(0.0, requires_grad=True).to(outputs.device)
    
    # Flatten everything and apply the mask
    masked_outputs = outputs[mask]
    masked_targets = targets[mask]
    
    return torch.nn.functional.cross_entropy(masked_outputs, masked_targets)
```

### 3. Training & Metric Logging
We train 4 models total: (Elman vs. CTRNN) $\times$ (Perceptual vs. Context). We train them using the Adam optimizer with early stopping. To ensure our results are reproducible and available for the final Technical Note, we log all training and validation metrics to a raw JSON file in the `metrics/` directory.

```python
def train_rnn(model, train_loader, val_loader, device, num_epochs=50, lr=1e-3, patience=5, use_early_stopping=True):
    """
    Trains an RNN with validation and early stopping.
    Only computes loss during the decision period (period == 2).
    """
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    
    best_val_loss = float('inf')
    patience_counter = 0

    # Store the initial weights just in case it completely fails to learn
    best_model_weights = copy.deepcopy(model.state_dict())
    
    history = {'train_loss': [], 'val_loss': []}
    
    for epoch in range(num_epochs):
        # --- TRAINING PHASE ---
        model.train()
        running_train_loss = 0.0
        train_batches = 0
        
        for obs, labels, periods, cohs, ctxs in train_loader:
            obs = obs.to(device)
            labels = labels.to(device)
            periods = periods.to(device)
            
            optimizer.zero_grad()
            
            # Forward pass
            outputs, _ = model(obs) # outputs shape: (Batch, Time, Classes)
            
            # Mask for decision period
            mask = (periods == 2)
            if mask.sum() > 0:
                masked_outputs = outputs[mask]
                masked_labels = labels[mask]
                
                loss = criterion(masked_outputs, masked_labels)
                loss.backward()
                optimizer.step()
                
                running_train_loss += loss.item()
                train_batches += 1
                
        avg_train_loss = running_train_loss / max(1, train_batches)
        history['train_loss'].append(avg_train_loss)
        
        # --- VALIDATION PHASE ---
        model.eval()
        running_val_loss = 0.0
        val_batches = 0
        
        with torch.no_grad():
            for obs, labels, periods, cohs, ctxs in val_loader:
                obs = obs.to(device)
                labels = labels.to(device)
                periods = periods.to(device)

                outputs, _ = model(obs)
                
                mask = (periods == 2)
                if mask.sum() > 0:
                    masked_outputs = outputs[mask]
                    masked_labels = labels[mask]
                    
                    loss = criterion(masked_outputs, masked_labels)
                    running_val_loss += loss.item()
                    val_batches += 1
                    
        avg_val_loss = running_val_loss / max(1, val_batches)
        history['val_loss'].append(avg_val_loss)
        
        # --- EARLY STOPPING CHECK ---
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            patience_counter = 0
            # Save model weights to keep the best version
            best_model_weights = copy.deepcopy(model.state_dict())
        else:
            patience_counter += 1
            
        # Print progress every 5 epochs
        if (epoch+1) % 5 == 0 or epoch == 0:
            print(f"Epoch [{epoch+1:03d}/{num_epochs}] | Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f}")
            
        if use_early_stopping and patience_counter >= patience:
            print(f"Early stopping triggered at epoch {epoch+1}! Best Val Loss: {best_val_loss:.4f}")
            break

    print("Training complete. Restoring best model weights...")
    # Load the best weights back into the model before returning
    model.load_state_dict(best_model_weights)
            
    return history, model
```

```python
# 7 input channels, 100 hidden units, 3 output choices (Fixate, Choice 1, Choice 2)
input_size = 7 
hidden_size = 100
output_size = 3

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using compute device: {device.type.upper()}")
if device.type == "cuda":
    print(f"   GPU Name: {torch.cuda.get_device_name(0)}")

# Directory Setup for Architecture Comparison
WEIGHTS_DIR = "model_weights"
METRICS_DIR = "metrics"
FIG_DIR = "figures"

os.makedirs(WEIGHTS_DIR, exist_ok=True)
os.makedirs(METRICS_DIR, exist_ok=True)
os.makedirs(FIG_DIR, exist_ok=True)
```

```python
print("--- Training Elman RNN on Perceptual Task ---")
per_model_elman = WrappedElman(input_size, hidden_size, output_size).to(device)
per_history_elman, per_model_elman = train_rnn(per_model_elman, per_train_loader, per_val_loader, device=device, num_epochs=50, patience=10)
per_model_elman.to('cpu')

# Save the best perceptual weights to disk
per_path_elman = os.path.join(WEIGHTS_DIR, f"elman_perceptual_best_h{hidden_size}.pth")
torch.save(per_model_elman.state_dict(), per_path_elman)
print(f"Saved Perceptual Model to {per_path_elman}\n")

print("\n--- Training Elman RNN on Context Task ---")
ctx_model_elman = WrappedElman(input_size, hidden_size, output_size).to(device)
ctx_history_elman, ctx_model_elman = train_rnn(ctx_model_elman, ctx_train_loader, ctx_val_loader, device=device, num_epochs=50, patience=10)
ctx_model_elman.to('cpu')

# Save the best context weights to disk
ctx_path_elman = os.path.join(WEIGHTS_DIR, f"elman_context_best_h{hidden_size}.pth")
torch.save(ctx_model_elman.state_dict(), ctx_path_elman)
print(f"Saved Context Model to {ctx_path_elman}\n")

print("\n--- Training CTRNN on Perceptual Task ---")
per_model_ctrnn = WrappedCTRNN(input_size, hidden_size, output_size).to(device)
per_history_ctrnn, per_model_ctrnn = train_rnn(per_model_ctrnn, per_train_loader, per_val_loader, device=device, num_epochs=50, patience=10)
per_model_ctrnn.to('cpu')

# Save the best perceptual weights to disk
per_path_ctrnn = os.path.join(WEIGHTS_DIR, f"ctrnn_perceptual_best_h{hidden_size}.pth")
torch.save(per_model_ctrnn.state_dict(), per_path_ctrnn)
print(f"Saved Perceptual Model to {per_path_ctrnn}\n")

print("\n--- Training CTRNN on Context Task ---")
ctx_model_ctrnn = WrappedCTRNN(input_size, hidden_size, output_size).to(device)
ctx_history_ctrnn, ctx_model_ctrnn = train_rnn(ctx_model_ctrnn, ctx_train_loader, ctx_val_loader, device=device, num_epochs=50, patience=10)
ctx_model_ctrnn.to('cpu')

# Save the best context weights to disk
ctx_path_ctrnn = os.path.join(WEIGHTS_DIR, f"ctrnn_context_best_h{hidden_size}.pth")
torch.save(ctx_model_ctrnn.state_dict(), ctx_path_ctrnn)
print(f"Saved Context Model to {ctx_path_ctrnn}\n")

# Compile all histories into one dictionary
all_metrics = {
    "elman_perceptual": per_history_elman,
    "elman_context": ctx_history_elman,
    "ctrnn_perceptual": per_history_ctrnn,
    "ctrnn_context": ctx_history_ctrnn
}

metrics_path = os.path.join(METRICS_DIR, "elman_vs_ctrnn_training_metrics.json")
with open(metrics_path, "w") as f:
    json.dump(all_metrics, f, indent=4)
    
print(f"Saved all training/validation metrics to {metrics_path}")
```

### 4. Information Geometry (PID) Extraction
Once trained, we pass a dedicated, unseen test set (2000 trials) through the networks. We freeze the weights and extract the hidden state $h(t)$ at *every single timestep*. Using the mean target coherence as our objective variable, we apply the Gaussian Minimal Mutual Information PID (MMI-PID) to map out how the network distributes information over time.

```python
def extract_hidden_trajectories(model, dataloader):
    """
    Runs the dataset through the model and extracts the hidden state, target, and context 
    at every timestep for every trial.
    """
    model.eval()
    all_hidden = []
    all_targets = []
    all_periods = []
    all_ctxs = []
    
    with torch.no_grad():
        for obs, labels, periods, cohs, ctxs in dataloader:
            # We bypass the final linear layer and get the raw hidden states
            # Standard nn.RNN returns (output, h_n). 'output' contains all timesteps.
            _, hidden_states = model(obs) 
            
            all_hidden.append(hidden_states.numpy())

            # Using mean coherence as our target!
            if cohs.ndim > 1:
                all_targets.append(cohs[:, 0].numpy().astype(float))
            else:
                all_targets.append(cohs.numpy().astype(float))
            # Using continuous noisy coherence as our target
            #noisy_evidence = obs[:, :, 1] - obs[:, :, 2]
            #all_targets.append(noisy_evidence.numpy().astype(float)) 
            
            # Using labels as our target!
            #all_targets.append(labels[:, -1].numpy().astype(float)) 

            all_periods.append(periods.numpy())

            all_ctxs.append(ctxs.numpy())
            
    # Concatenate across batches
    H = np.concatenate(all_hidden, axis=0)      # Shape: (Trials, Timesteps, Hidden_Size)
    Y = np.concatenate(all_targets, axis=0)     # Shape: (Trials,)
    P = np.concatenate(all_periods, axis=0)     # Shape: (Trials, Timesteps)
    C = np.concatenate(all_ctxs, axis=0)        # Shape: (Trials,)
    
    return H, Y, P, C
```

```python
print("Extracting hidden state trajectories for the Test Sets...")

# Load the dedicated Test Sets (never seen during training or validation!)
per_test_loader = load_mante_data(f'{BASE_PATH}/perceptual/test_01.npz', batch_size=BATCH_SIZE, shuffle=False, subsample_step=SUBSAMPLE_STEP)
ctx_test_loader = load_mante_data(f'{BASE_PATH}/context/test_01.npz', batch_size=BATCH_SIZE, shuffle=False, subsample_step=SUBSAMPLE_STEP)

# Load models from savepoints (optional)
per_model_elman = WrappedElman(input_size, hidden_size, output_size)
per_model_elman.load_state_dict(torch.load(f"{WEIGHTS_DIR}/elman_perceptual_best_h{hidden_size}.pth", weights_only=True))
context_model_elman = WrappedElman(input_size, hidden_size, output_size)
context_model_elman.load_state_dict(torch.load(f"{WEIGHTS_DIR}/elman_context_best_h{hidden_size}.pth", weights_only=True))

perceptual_model_ctrnn = WrappedCTRNN(input_size, hidden_size, output_size)
perceptual_model_ctrnn.load_state_dict(torch.load(f"{WEIGHTS_DIR}/ctrnn_perceptual_best_h{hidden_size}.pth", weights_only=True))
context_model_ctrnn = WrappedCTRNN(input_size, hidden_size, output_size)
context_model_ctrnn.load_state_dict(torch.load(f"{WEIGHTS_DIR}/ctrnn_context_best_h{hidden_size}.pth", weights_only=True))

# Extract Trajectories
H_per_elman, Y_per_elman, P_per_elman, C_per_elman = extract_hidden_trajectories(per_model_elman, per_test_loader)
H_ctx_elman, Y_ctx_elman, P_ctx_elman, C_ctx_elman = extract_hidden_trajectories(context_model_elman, ctx_test_loader)

H_per_ctrnn, Y_per_ctrnn, P_per_ctrnn, C_per_ctrnn = extract_hidden_trajectories(perceptual_model_ctrnn, per_test_loader)
H_ctx_ctrnn, Y_ctx_ctrnn, P_ctx_ctrnn, C_ctx_ctrnn = extract_hidden_trajectories(context_model_ctrnn, ctx_test_loader)

print(f"Perceptual Trajectories Shape: {H_per_elman.shape}")
print(f"Context Trajectories Shape:    {H_ctx_elman.shape}")
```

```python
# Run the analysis (Warning: This might take a few minutes on CPU!)
print("Analyzing Perceptual Elman Network...")
pid_elman_perceptual = gaussian_pid_rnn_gpu(
    H=H_per_elman,
    Y=Y_per_elman, 
    n_bip=200,
    seed=42,
    reg=1e-5
)

print("Analyzing Context Elman Network...")
pid_elman_context = gaussian_pid_rnn_gpu(
    H=H_ctx_elman,
    Y=Y_ctx_elman, 
    n_bip=200,
    seed=42,
    reg=1e-5
)

print("Analyzing Perceptual CTRNN Network...")
pid_ctrnn_perceptual = gaussian_pid_rnn_gpu(
    H=H_per_ctrnn,
    Y=Y_per_ctrnn, 
    n_bip=200,
    seed=42,
    reg=1e-5
)

print("Analyzing Context CTRNN Network...")
pid_ctrnn_context = gaussian_pid_rnn_gpu(
    H=H_ctx_ctrnn,
    Y=Y_ctx_ctrnn, 
    n_bip=200,
    seed=42,
    reg=1e-5
)

print("Succesfully calculated PID metrics")
```

### 5. Evaluating Representational Capacity (Visualization)
We plot the Synergy, Redundancy, Unique, and Total information tracks over time. 
If our hypothesis holds, the CTRNN will show a stable, well-behaved information geometry (with Synergy dominating the Context task). The Elman RNN, if it fails to converge smoothly due to noise, will likely exhibit collapsed or chaotic information profiles.

```python
# Dynamically find the End of Stimulus index (same across all models/tasks)
stim_end_idx = np.where(P_ctx_elman[0] == 1)[0][-1] 

# Define the grid
fig, axes = plt.subplots(2, 2, figsize=(18, 12), sharey=True, sharex=True)

# Add an overarching title to the entire figure specifying the hidden size
fig.suptitle(f"Information Geometry: Architecture vs. Cognitive Demand (Hidden Size: {hidden_size})", 
             fontsize=20, fontweight='bold', y=0.96)

def plot_pid_ax(ax, data, title):
    """Helper function to plot PID lines on a specific subplot axis."""
    # data is of shape (timesteps, [total, red, u1, u2, syn])
    time_axis = np.arange(len(data[:, 0]))
    
    ax.plot(time_axis, data[:, 4], label='Synergy', color='#54A24B', linewidth=3)
    ax.plot(time_axis, data[:, 1], label='Redundancy', color='#4C78A8', linewidth=3)
    ax.plot(time_axis, data[:, 2], label='Unique 1', color='#F5A300', linewidth=1.5, alpha=0.8)
    ax.plot(time_axis, data[:, 3], label='Unique 2', color='#E618F5', linewidth=1.5, alpha=0.8)
    ax.plot(time_axis, data[:, 0], label='Total MI', color='#020202', linewidth=1.5, alpha=0.8, linestyle=':')
    
    # Mark the End of Stimulus
    ax.axvline(stim_end_idx, color='red', linestyle='--', label='Decision time')
    ax.axvspan(0, stim_end_idx, color='gray', alpha=0.1)
    
    ax.set_title(title, fontsize=15, fontweight='bold', pad=10)
    ax.grid(alpha=0.3)
    
    # Only show axis labels on the outer edges to keep it clean
    if ax.get_subplotspec().is_first_col():
        ax.set_ylabel('Information (Bits)', fontsize=13)
    if ax.get_subplotspec().is_last_row():
        ax.set_xlabel('Timesteps (Subsampled)', fontsize=13)

# --- Row 1: Elman RNN ---
plot_pid_ax(axes[0, 0], pid_elman_perceptual, "Elman RNN | Perceptual Task")
plot_pid_ax(axes[0, 1], pid_elman_context, "Elman RNN | Context Task")

# --- Row 2: CTRNN ---
plot_pid_ax(axes[1, 0], pid_ctrnn_perceptual, "CTRNN | Perceptual Task")
plot_pid_ax(axes[1, 1], pid_ctrnn_context, "CTRNN | Context Task")

# Add a single master legend to the first plot
axes[0, 0].legend(loc='upper left', fontsize=11, framealpha=0.9)

# Adjust spacing so the main title doesn't overlap with the subplot titles
plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.savefig(os.path.join(FIG_DIR, "elman_vs_ctrnn_pid_geometry.png"), dpi=300, bbox_inches='tight')
plt.show()
```
