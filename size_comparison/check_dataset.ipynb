```python
import numpy as np
import matplotlib.pyplot as plt

CDMdata = np.load(
    r"E:\TUM\3sem\NeuroAI\project\data_generation\CDM\seed42\context\train.npz",
    allow_pickle=True
)

obs = CDMdata["observations"]
labels = CDMdata["labels"]
coh = CDMdata["coherences"]
ctx = CDMdata["contexts"]
periods = CDMdata["trial_periods"]

i = 0  # pick trial index

print("coherence:", coh[i])
print("context:", ctx[i])

T = obs.shape[1]

plt.figure(figsize=(14, 6))

# ---- plot a few observation channels ----
for ch in range(min(5, obs.shape[2])):  # first 5 channels
    plt.plot(obs[i, :, ch], label=f"ch {ch}")

# ---- overlay task periods ----
plt.plot(periods[i] * 0.5, label="period (scaled)", alpha=0.6)

plt.title(f"Trial {i} | coh={coh[i]:.2f} | ctx={ctx[i]}")
plt.xlabel("time")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(4, 1, figsize=(14, 8), sharex=True)

i = 2
stim_period = periods[i] == 1
print("coherence:", coh[i])
# observations (sum of first stimulus channels for clarity)
ax[0].plot(obs[i, :, 1], label="stim1_mod1")
ax[0].axhline(np.mean(obs[i, stim_period, 1]), linestyle="--",
              label=f"mean stim1_mod1={np.mean(obs[i, stim_period, 1]):.3f}")
ax[0].plot(obs[i, :, 3], label="stim1_mod2")
ax[0].axhline(np.mean(obs[i, stim_period, 3]), linestyle="--",
              label=f"mean stim1_mod2={np.mean(obs[i, stim_period, 3]):.3f}")
ax[0].set_title("Stimulus channels")
ax[0].legend()

ax[1].plot(obs[i, :, 2], label="stim2_mod1")
ax[1].axhline(np.mean(obs[i, stim_period, 2]), linestyle="--",
              label=f"mean stim2_mod1={np.mean(obs[i, stim_period, 2]):.3f}")
ax[1].plot(obs[i, :, 4], label="stim2_mod2")
ax[1].axhline(np.mean(obs[i, stim_period, 4]), linestyle="--",
              label=f"mean stim2_mod2={np.mean(obs[i, stim_period, 4]):.3f}")
ax[1].set_title("Stimulus channels")
ax[1].legend()

# labels (decision target)
ax[2].plot(labels[i], color="black")
ax[2].set_title("Decision label")

# task period
ax[3].plot(periods[i], color="gray")
ax[3].set_title("Fixation / Stimulus / Decision")
ax[3].set_yticks([0,1,2])

plt.tight_layout()
plt.show()

stim_period = periods[i] == 1

print("mod1 difference:",
      np.mean(obs[i, stim_period, 1]) -
      np.mean(obs[i, stim_period, 2]))

print("mod2 difference:",
      np.mean(obs[i, stim_period, 3]) -
      np.mean(obs[i, stim_period, 4]))
```

```python
import numpy as np

PDMdata = np.load(
    r"E:\TUM\3sem\NeuroAI\project\data_generation\PDM\seed42\perceptual\test_uniform.npz"
)

print(PDMdata.files)

obs = PDMdata["observations"]
labels = PDMdata["labels"]
coh = PDMdata["coherences"]
ctx = PDMdata["contexts"]
periods = PDMdata["trial_periods"]

print("obs shape:", obs.shape)
print("coh range:", coh.min(), coh.max())
print("unique contexts:", np.unique(ctx))
```

```python
fig, ax = plt.subplots(4, 1, figsize=(14, 8), sharex=True)

i = 0

trial_coh = coh[i]
mean_coh = np.mean(coh)

stim_period = periods[i] == 1
fig.suptitle(
    f"Trial {i} | coherence={trial_coh:.3f} | dataset mean coherence={mean_coh:.3f}",
    fontsize=14
)
print("coherence:", coh[i])
ax[0].plot(obs[i, :, 1], label="stim1_mod1")
ax[0].axhline(np.mean(obs[i, stim_period, 1]), linestyle="--",
              label=f"mean stim1_mod1={np.mean(obs[i, stim_period, 1]):.3f}")
ax[0].plot(obs[i, :, 3], label="stim1_mod2")
ax[0].set_title("Stimulus channels")
ax[0].legend()

ax[1].plot(obs[i, :, 2], label="stim2_mod1")
ax[1].axhline(np.mean(obs[i, stim_period, 2]), linestyle="--",
              label=f"mean stim2_mod1={np.mean(obs[i, stim_period, 2]):.3f}")
ax[1].plot(obs[i, :, 4], label="stim2_mod2")
ax[1].set_title("Stimulus channels")
ax[1].legend()

ax[2].plot(labels[i], color="black")
ax[2].set_title("Decision label")

ax[3].plot(periods[i], color="gray")
ax[3].set_title("Fixation / Stimulus / Decision")
ax[3].set_yticks([0, 1, 2])

plt.tight_layout()
plt.show()



m1 = np.mean(obs[i, stim_period, 1])
m2 = np.mean(obs[i, stim_period, 2])

print("coherence =", coh[i])
print("mean stim1 =", m1)
print("mean stim2 =", m2)
print("difference =", m1 - m2)
```
